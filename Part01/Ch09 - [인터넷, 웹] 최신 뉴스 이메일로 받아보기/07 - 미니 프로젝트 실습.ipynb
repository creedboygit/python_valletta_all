# 웹페이지 저장

```python
import urllib
from urllib import request
```

```python
# 뉴스 속보 페이지를 html 파일로 저장
def save_to_html(url):
    with urllib.request.urlopen(url) as s:  # 웹 페이지 리소스 객체 생성
        with open('breaking_news.html', 'wb') as f:
            f.write(s.read())  # 리소스 내용을 읽어서 html 파일로 저장
```

```python
save_to_html(url)
```

# 뉴스 속보 크롤링

```python
from bs4 import BeautifulSoup
import requests

# 헤더 추가
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

# 서버 응답 확인
response = requests.get(url, headers=headers)
```

```python
response
```

```python
# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# print(soup.prettify())
print(soup.prettify())
```

```python
# 페이지 제목 크롤링
page_title = soup.title.string
print(page_title)
```

```python
# 공백 제거
import re

regex = re.compile(r'[\n\r\t]')
page_title = regex.sub(' ', page_title)
```

```python
page_title
```

```python
# 페이지 상단 텍스트 크롤링
main_contents = soup.find("dl", attrs={"class": "type04"}).get_text()
print(main_contents)
```

```python
# 페이지 하단 텍스트 크롤링
sub_contents = soup.find("ul", attrs={"class": "type02"}).get_text()
print(sub_contents)
```

# 이메일 보내기

```python
import smtplib
from email.mime.multipart import MIMEMultipart

msg = MIMEMultipart()
```

```python
# 이메일 송수신자 설정
msg['From'] = 'polozhzh@gmail.com'
msg['To'] = 'polozhzh@gmail.com'
```

```python
# 날짜 설정
from email.utils import formatdate

msg['Date'] = formatdate(localtime=True)  # 현재 지역에 맞는 날짜
msg['Date']
```

```python
# 이메일 제목/본문 작성
from email.header import Header
from email.mime.text import MIMEText

msg['Subject'] = Header(s=page_title, charset='utf-8')
body = MIMEText(main_contents + sub_contents, _charset='utf-8')
```

```python
# 메일 본문 추가
msg.attach(body)
```

```python
msg
```

```python
print(msg)
```

```python
# 파일 첨부
import os
from email.mime.base import MIMEBase
from email.encoders import encode_base64

files = list()
files.append('breaking_news.html')

for f in files:
    part = MIMEBase('application', 'octet-stream')
    part.set_payload(open(f, 'rb').read())
    encode_base64(part)
    part.add_header('Content-Disposition', 'attachment; filename="%s"' % os.path.basename(f))
    msg.attach(part)  # 파일 첨부
```

```python
# 메일 발송
import getpass

mail_server = smtplib.SMTP_SSL('smtp.gmail.com')  # smtp 서버를 사용한 메시지 발송
mail_server.login('polozhzh@gmail.com', getpass.getpass())  # 본인 계정과 비밀번호 사용 (Gmail : 앱 비밀번호)
mail_server.send_message(msg)
mail_server.quit()
```

# 이메일 수신 확인

```python
import poplib

# 암호화된 소켓 연결 (기본 995 포트)
server = poplib.POP3_SSL('pop.googlemail.com', port=995)

# google id 입력
server.user('polozhzh')
```

```python
# password 입력
import getpass

server.pass_(getpass.getpass())
```

```python
server.stat()
```

```python
import email
from email.header import decode_header, make_header

# 가장 최근 메시지 확인
recent_no = server.stat()[0]
```

```python
recent_no
```

```python
# 메시지 가져오기
server.retr(recent_no)
```

```python
# 줄바꿈으로 메시지 내용 조인
raw_email = b'\n'.join(server.retr(recent_no)[1])
```

```python
raw_email
```

```python
# 메시지 객체 생성
message = email.message_from_bytes(raw_email)

message
```

```python
print(message)
```

```python
# 송신자 확인
fr = message.get('From')
print(fr)
```

```python
# 제목 확인
subject = make_header(decode_header(message.get('Subject')))
subject
```

```python
print(subject)
```

```python
# 본문 확인
attachments = []

if message.is_multipart():
    # 멀티파트라면 여러 개로 나누어진 메시지를 하나씩 처리
    for part in message.walk():
        ctype = part.get_content_type()
        cdispo = str(part.get('Content-Disposition'))
        print(ctype)

        # 컨텐츠에 첨부 파일이 있다면
        if 'attachment' in cdispo:
            name = part.get_filename()
            data = part.get_payload(decode=True)
            f = open(name, 'wb')
            f.write(data)
            f.close()
            attachments.append(name)

        # 컨텐츠가 text/plain이고, 첨부파일이 없다면
        if ctype == 'text/plain' and 'attachment' not in cdispo:
            body = part.get_payload(decode=True)  # 메시지 내용 추출
else:
    # 싱글 파트라면
    body = message.get_payload(decode=True)  # 메시지 내용 추출
```

```python
body = body.decode('utf-8')

print(f'보낸사람:{fr}')
print(f'제목:{subject}')
print(f'내용:{body}')
print(f'첨부파일:{attachments}')
```

```python
# 첨부파일 열기
webbrowser.open('file://' + os.path.realpath(attachments[0]))

os.path.realpath(attachments[0])
```
