### base64 디코딩

```python
# 기사 이미지 디코딩
import base64
file_base64 = image[0]
```

```python
path = "news/image.jpg"
with open(path, 'wb') as f:
    decoded_data = base64.decodebytes(file_base64)
    f.write(decoded_data)
```

```python
# 이미지 확인
from PIL import Image
img = Image.open(path)
img
```

```python
# 기사 본문 디코딩
file_base64 = article[0]
decoded_data = base64.decodebytes(file_base64)
article = decoded_data.decode('utf-8')
```

```python
print(article)
```

# 키워드 추출

```python
import collections
import textwrap
import re
```

```python
# 줄바꿈 정렬
article_align = textwrap.fill(article, width=50)
print(article_align)
```

```python
# 단어 추출
words = re.findall(r'\w+', article_align)
words
```

```python
# 빈도수 산출
counter = collections.Counter(words)
counter
```

```python
# 키워드 추출
print(counter.most_common(5))
```

```python
keywords = counter.most_common(5)[1:]
keywords
```

```python
from IPython.display import Image
Image(filename=path, width=300)
```

```python
keys = ['#' + elem[0] for elem in keywords]
keys = ' '.join(keys)
keys
```

# html 파일로 저장
