```python
import base64
```

```python
string = 'Life is too short, We need Python !'
```

```python
encoded = base64.b64encode(string)
```

```python
# ascii 인코딩
bstring = string.encode('ascii')
print(bstring)
```

```python
# base64 인코딩
encoded = base64.b64encode(bstring)
```

```python
print(encoded)
```

```python
# base64 디코딩
decoded = base64.decodebytes(encoded)
```

```python
print(decoded)
```

```python
# ascii 디코딩
decoded_str = decoded.decode('ascii')
print(decoded_str)
```
